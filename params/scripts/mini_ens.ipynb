```python
oaats = {}
oaats['default']=[np.nan]
oaats['kmax']=[2e-8/2**x for x in 1+np.arange(6)]
oaats['hksat_adjustfactor']=[0.9,1.1]
oaats['medlynslope']=[3,4,5,6]
```

```python
prefix = 'miniens_oaat'
first = 0
i = -1
for v in oaats:
    vals = oaats[v]
    nx = len(vals)
    for j in range(nx):
        i +=1
        pftfile = "../paramfiles/spinme/"+prefix+str(first+i).zfill(4)+".nc"
        print(pftfile)
        cmd = 'cp '+basepftfile+' '+pftfile
        os.system(cmd)
        if v!='default':
            dset = netCDF4.Dataset(pftfile,'r+')
            dset[v][:]=oaats[v][j]
            dset.close()
```
